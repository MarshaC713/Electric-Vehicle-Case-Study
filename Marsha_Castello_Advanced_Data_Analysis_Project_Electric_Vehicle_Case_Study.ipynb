##### Analysis of the values of the different quartiles to find the IQR. 
#####  The IQR (interquartile range) is the difference between the 75th (q3) 
#####  and 25th (q1) percentile of the data. 
##### It is a measure of the dispersion similar to standard deviation or variance,
###### but is much more robust against outliers.
##### to do this I first confirm and determine the values of each:

```python
# The 75th(q3) and 25th(q1) quartiles are 410, and 265 respectively 
# as returned by the describe() function

q3 = df['Electric Range'].quantile(0.75)
q1 = df['Electric Range'].quantile(0.25)
print(q3)
print(q1)
```

```python
# Finding the IQR (interquartile range):

IQR = q3 - q1
print(IQR)
```

##### A commonly used rule says that a data point is an outlier if it is more than 1.5 * IQR above the third quartile or below the first quartile.
##### In other words low outliers are below q1 - (IQR * 1.5).
##### High outliers are above q3 + (IQR * 1.5)

```python
# Electric Ranges below 47.5 are outliers

lower_limit = q1 - (IQR * 1.5)
print(lower_limit)
```

```python
# Electric Ranges above 627.5 are outliers
# So an Electric Range of 640 is higher than 627.5 and
# definitely an outlier.
upper_limit = q3 + (IQR * 1.5)
print(upper_limit)
```

```python
# Detecting the outliers:
df_outliers = np.where(df['Electric Range'] > upper_limit, True,
                       np.where(df['Electric Range'] < lower_limit, True, False))
```

```python
# Excluding the outliers to confirm mean (average) value:
df_without_outliers = df.loc[~(df_outliers)]
```

```python
# Removing the outliers to confirm and display the average electric range
# in a box plot diagram:

a=sns.boxplot( y='Electric Range', data=df_without_outliers)
a
```
